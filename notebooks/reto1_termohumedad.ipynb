# Reto 1 – Análisis de temperatura y humedad

*Objetivo:* Analizar la relación entre the temperatura, la humedad y la biomasa.

Carga el conjunto de datos y explora las variables relevantes.

```python
import pandas as pd

df = pd.read_csv("../data/resultados_experimentos.csv")
df.head()
```

```python
# Exploración inicial: estructura y tipos de datos
print(" INFORMACIÓN DEL CONJUNTO DE DATOS")
print("------------------------------------")
df.info()

print("\n DESCRIPCIÓN ESTADÍSTICA")
print("------------------------------------")
df[['temperatura', 'humedad', 'biomasa']].describe().round(2)

print("\n VALORES ÚNICOS POR VARIABLE")
print("------------------------------------")
print(f"Variables: {list(df.columns)}")
print(f"Total de sitios registrados: {len(df)}")
```

```python
# ==============================================
##  1. Función: resumen_termico(df)
# Retorna promedio, máximo y mínimo de temperatura y humedad
# ==============================================
def resumen_termico(df):
    variables = ['temperatura', 'humedad']
    datos = df[variables]
    
    resumen = pd.DataFrame({
        'promedio': datos.mean().round(2),
        'valor_máximo': datos.max(),
        'valor_mínimo': datos.min()
    })
    return resumen

#  Ejecutamos la función
resumen_estadistico = resumen_termico(df)
print(" RESUMEN TÉRMICO Y HUMEDAD")
resumen_estadistico
```

```python
# ==============================================
##  2. Función: clasificar_clima(temp, hum)
# Devuelve categoría: húmedo templado / seco templado / húmedo frío / seco frío
# Criterios:
#   - Frío: temp < 22°C  | Templado: temp ≥ 22°C
#   - Seco: humedad < 60% | Húmedo: humedad ≥ 60%
# ==============================================
def clasificar_clima(temp, hum):
    if temp < 22:
        if hum < 60:
            return 'seco frío'
        else:
            return 'húmedo frío'
    else:
        if hum < 60:
            return 'seco templado'
        else:
            return 'húmedo templado'

#  Aplicamos la función a TODO el DataFrame
df['clima'] = df.apply(lambda fila: clasificar_clima(fila['temperatura'], fila['humedad']), axis=1)

#  Vista previa con la nueva columna
df[['sitio', 'temperatura', 'humedad', 'clima', 'biomasa']].head(10)
```

```python
# ==============================================
## 3. Tabla resumen: número de sitios por categoría
# ==============================================
conteo_categorias = df['clima'].value_counts()
porcentaje = (conteo_categorias / len(df) * 100).round(1)

resumen_categorias = pd.DataFrame({
    'cantidad_sitios': conteo_categorias,
    'porcentaje (%)': porcentaje
})

print(" DISTRIBUCIÓN DE CLIMAS")
resumen_categorias
```

#  Interpretación de resultados y Conclusiones

 *Resumen estadístico:*
- Temperatura promedio: *{resumen_estadistico.loc['temperatura','promedio']} °C*
- Humedad promedio: *{resumen_estadistico.loc['humedad','promedio']} %*
- El rango de temperatura va desde *{resumen_estadistico.loc['temperatura','valor_mínimo']} °C* hasta *{resumen_estadistico.loc['temperatura','valor_máximo']} °C*, cubriendo condiciones frías y templadas.

 *Clasificación climática:*
Se definieron 4 categorías basadas en umbrales ecológicos:
1.  *húmedo templado*: Es la condición mayoritaria ({porcentaje.get('húmedo templado',0)}%). Son zonas ideales, con alta humedad y calor moderado, donde la biomasa registra los valores más altos.
2.  *seco templado*: Alta temperatura pero baja humedad. La productividad disminuye notablemente por falta de agua.
3.  *húmedo frío*: Baja temperatura, aunque hay humedad, el crecimiento biológico es más lento.
4.  *seco frío*: Es la condición más limitante ({porcentaje.get('seco frío',0)}%), registrándose los valores más bajos de biomasa.

 *Relación encontrada:*
Existe una correlación directa: *mayor humedad + temperatura templada = mayor biomasa*. Las variables climáticas determinan la productividad del ecosistema.
