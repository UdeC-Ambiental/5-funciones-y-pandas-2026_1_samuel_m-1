```python
def clasificar_ph(ph):
    if ph < 6.5:
        return 'Ácido'
    elif ph <= 8.5:
        return 'Neutro'
    else:
        return 'Básico'

print(clasificar_ph(6.0))
print(clasificar_ph(7.4))
print(clasificar_ph(9.1))
```

```python
def promedio(lista):
    suma = 0
    for valor in lista:
        suma += valor
    return suma / len(lista)

temperaturas = [18.2, 21.5, 19.8, 22.3]
print('Promedio de temperatura:', round(promedio(temperaturas), 2))
```

```python
def analizar_temperaturas(lista_temperatura):
    for t in lista_temperatura:
        if t > 25:
            print(t, '→ Alta temperatura')
        else:
            print(t, '→ Temperatura normal')

# Lista
temperaturas = [18.5, 22.4, 27.1]
analizar_temperaturas(temperaturas)

# Tupla
mediciones = (19.0, 25.3, 29.5)
analizar_temperaturas(mediciones)
```

```python
def mostrar_calidad(diccionario):
    for sitio, ph in diccionario.items():
        print(sitio, '→', clasificar_ph(ph))

sitios = {'Río Subachoque': 7.2, 'Quebrada Cune': 6.1, 'Río Bahamón': 9.0}
mostrar_calidad(sitios)

# --------------------------
import pandas as pd

def evaluar_calidad(df):
    df['estado'] = df['pH'].apply(clasificar_ph)
    return df

datos = pd.DataFrame({
    'sitio': ['Río Subachoque', 'Río Villeta', 'Quebrada Cune'],
    'pH': [7.2, 8.7, 6.0]
})

df_resultado = evaluar_calidad(datos)
print(df_resultado)
```
